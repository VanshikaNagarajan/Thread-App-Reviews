```python
import pandas as pd
threads_S_result = pd.read_csv('sentiment_check_results.csv')

import matplotlib.pyplot as plt
rating_sentiment = threads_S_result.groupby('sentiment')['rating'].mean().reset_index()
```

```python
plt.figure(figsize = (8,4))
plt.bar(rating_sentiment['sentiment'], rating_sentiment['rating'], color = ['yellow', 'red'])
plt.xlabel('user rating')
plt.ylabel('sentiment')
plt.title('comparison between the user rating and review sentiment analysis')
plt.show()
```
